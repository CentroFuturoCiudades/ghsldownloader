# Example usage

To use `ghsldownloader` in a project:

```python
from ghsldownloader.ghsldownloader import download_ghsl
from osgeo import gdal
import matplotlib.pyplot as plt
from shapely import box
```

```python
bbox = box(-9497448.733333638, 3080298.9443988125, -9372885.4431954 , 3176676.5574316736)
```

```python
download_ghsl("./MTY", products=('POP', 'BUILT_S', 'LAND'), epochs=(2020,), crs=54009, resolution=100, extent="bbox", bbox=bbox, prefix="MTY_")
```

```python
ds = gdal.Open("MTY/warped_MTY_BUILT_S_2020_54009_100.tif")
band = ds.GetRasterBand(1)
arr = band.ReadAsArray()
del ds
arr.shape
```

```python
plt.imshow(arr);
```
